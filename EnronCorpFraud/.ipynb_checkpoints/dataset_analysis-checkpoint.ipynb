#### Looks like all these messages share the same structure, which is quite convenient. It means I can split the text by newline ('\n\') and then simply slice the resulting list to access specific elements of each email. Then I can assign them to new dataframe columns and clean them up for further exploration.

```python
dataframe['text'] = dataframe['message'].apply(lambda message: message.split('\n')[15:])
```

```python
dataframe.head(5)
```

```python
dataframe['date'] = dataframe['message'].apply(lambda message: message.split('\n')[1][6:])
dataframe['sender'] = dataframe['message'].apply(lambda message: message.split('\n')[2][6:])
dataframe['recipients'] = dataframe['message'].apply(lambda message: message.split('\n')[3][3:].split(','))
dataframe['subject'] = dataframe['message'].apply(lambda message: message.split('\n')[4][8:])
```
